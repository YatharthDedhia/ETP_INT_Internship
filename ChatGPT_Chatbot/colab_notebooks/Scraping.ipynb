```python
# importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_titles(soup):
    div_elements = soup.find_all('div', class_='_4rR01T')
    titles_list = []
    for div in div_elements:
        text = div.get_text(strip=True)
        titles_list.append(text)
    return titles_list
```

```python
def get_price(soup):
    div_elements = soup.find_all('div', class_='_30jeq3')
    titles_list = []
    for div in div_elements:
        text = div.get_text(strip=True)
        titles_list.append(text)
    return titles_list
```

```python
def get_prod_urls(soup):
    a_elements = soup.find_all('a', class_='_1fQZEK')

    url_list = []
    for a in a_elements:
        href = a.get('href')
        if href:
            url_list.append("https://www.flipkart.com"+href)

    return url_list
```

```python
product_name = "pixel"
```

```python
all_titles=[]
all_urls=[]
all_prices=[]
for page_no in range(1,10):
  URL = "https://www.flipkart.com/search?q="+product_name+"&page="+str(page_no)
  webpage = requests.get(URL).text
  soup = BeautifulSoup(webpage, "lxml")
  all_titles.extend(get_titles(soup))
  all_urls.extend(get_prod_urls(soup))
  # all_prices.extend(get_price(soup))
```

```python
df = pd.DataFrame({'Title': all_titles, 'URL': all_urls})
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create lists to store the extracted data
product_names = []
product_prices = []
product_ratings = []
product_specifications = []
product_urls = []

for i in range(1, 10):
    # Send a GET request to the Flipkart page
    url = 'https://www.flipkart.com/search?q=pixel' + '&page=' + str(i)
    response = requests.get(url)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the elements with class "_1AtVbE col-12-12"
    product_elements = soup.find_all('div', class_='_1AtVbE col-12-12')

    # Iterate over the product elements and extract the required data
    for element in product_elements:
        # Extract product name from img alt attribute
        product_name_element = element.find('img', class_='_396cs4')
        product_name = product_name_element['alt'] if product_name_element else None

        # Extract product price
        product_price_element = element.find('div', class_='_30jeq3 _1_WHN1')
        product_price = product_price_element.text if product_price_element else None

        # Extract product rating
        product_rating_element = element.find('div', class_='_3LWZlK')
        product_rating = product_rating_element.text if product_rating_element else None

        # Check if all details are present before appending to the lists
        if product_name and product_price and product_rating:
            product_names.append(product_name)
            product_prices.append(product_price)
            product_ratings.append(product_rating)

            # Extract product specifications
            product_specification_element = element.find('ul', class_='_1xgFaf')
            product_specification = product_specification_element.text if product_specification_element else None
            product_specifications.append(product_specification)

            # Extract product URL
            product_url_element = element.find('a', class_='_1fQZEK')
            product_url = 'https://www.flipkart.com' + product_url_element['href'] if product_url_element else None
            product_urls.append(product_url)

# Create a pandas dataframe from the extracted data
data = {
    'Product Name': product_names,
    'Price': product_prices,
    'Rating': product_ratings,
    'Specifications': product_specifications,
    'Product URL': product_urls
}

df = pd.DataFrame(data)

# Remove rows with None values
df = df.dropna()

# Reset the row index
df = df.reset_index(drop=True)

# Display the dataframe
df
```
